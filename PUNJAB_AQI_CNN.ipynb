```python
# Display column names to check for any issues
print(hourly_data.columns)

# Strip any leading/trailing whitespace characters from the column names
hourly_data.columns = hourly_data.columns.str.strip()

# Check if 'date' column exists and convert to datetime
if 'date' in hourly_data.columns:
    # Try to infer the format automatically, handling errors individually
    hourly_data['date'] = pd.to_datetime(hourly_data['date'], errors='coerce')
    # Set date as index
    hourly_data.set_index('date', inplace=True)
    print(hourly_data.head())
else:
    print("Error: 'date' column not found in the dataset.")
```

```python


# Handle missing values (if any)
hourly_data.fillna(method='ffill', inplace=True)
```

```python
# Import the necessary library
from sklearn.preprocessing import MinMaxScaler

# Normalize the features
scaler = MinMaxScaler()
# Extract numerical features for scaling, EXCLUDING 'latitude', 'longitude', and column names
numerical_features = hourly_data.drop(columns=['latitude', 'longitude']).select_dtypes(include=['number'])
scaled_data = scaler.fit_transform(numerical_features)

# Create a DataFrame from scaled data
scaled_data = pd.DataFrame(scaled_data, columns=numerical_features.columns, index=hourly_data.index)

# Add back latitude and longitude
scaled_data['latitude'] = hourly_data['latitude']
scaled_data['longitude'] = hourly_data['longitude']

print(scaled_data.head())
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout, Flatten

# Function to create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

# Prepare data for CNN-LSTM
SEQ_LENGTH = 24  # Using past 24 hours to predict the next hour
features = scaled_data.drop(columns=['latitude', 'longitude']).values
X, y = create_sequences(features, SEQ_LENGTH)

# Split the data into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Reshape data for CNN input
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2], 1))
```

```python
# Build the CNN-LSTM model
model = Sequential([
    Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(SEQ_LENGTH, X_train.shape[2])),
    MaxPooling1D(pool_size=2),
    LSTM(50, activation='relu', return_sequences=True),
    LSTM(50, activation='relu'),
    Dense(50, activation='relu'),
    Dropout(0.2),
    Dense(y_train.shape[1])
])

model.compile(optimizer='adam', loss='mse')
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

```python
def calculate_aqi(pm10, pm2_5, co, no2, so2, o3, dust):
    weight_pm10 = 0.25
    weight_pm2_5 = 0.25
    weight_co = 0.1
    weight_no2 = 0.15
    weight_so2 = 0.1
    weight_o3 = 0.1
    weight_dust = 0.05

    aqi = (weight_pm10 * pm10 + weight_pm2_5 * pm2_5 + weight_co * co +
           weight_no2 * no2 + weight_so2 * so2 + weight_o3 * o3 + weight_dust * dust)
    return aqi

# Apply AQI calculation
scaled_data['AQI'] = scaled_data.apply(lambda row: calculate_aqi(row['pm10'], row['pm2_5'], row['carbon_monoxide'],
                                                                 row['nitrogen_dioxide'], row['sulphur_dioxide'],
                                                                 row['ozone'], row['dust']), axis=1)
```

```python
def predict_aqi(model, data, seq_length):
    predictions = []
    current_seq = data[-seq_length:].values
    for _ in range(14 * 24):  # Predict for 14 days (24 hours each day)
        current_seq_scaled = scaler.transform(current_seq)
        current_seq_scaled = current_seq_scaled.reshape((1, seq_length, current_seq_scaled.shape[1]))
        pred = model.predict(current_seq_scaled)
        pred_unscaled = scaler.inverse_transform(pred)
        predictions.append(pred_unscaled[0])
        current_seq = np.append(current_seq[1:], [pred_unscaled[0]], axis=0)
    return predictions
```
