```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import files
uploaded = files.upload()

# Load the image using OpenCV
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)
```

```python
cv2_imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
inverted_image = cv2.bitwise_not(gray_image)
cv2_imshow(inverted_image)
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
cv2_imshow(blurred_image)
```

```python
edges = cv2.Canny(gray_image, 100, 200)
cv2_imshow(edges)
```

```python
edges = cv2.Canny(gray_image, 100, 200)
cv2_imshow(edges)
```

```python
cv2.imwrite('inverted_image.png', inverted_image)
cv2.imwrite('blurred_image.png', blurred_image)
cv2.imwrite('edges_image.png', edges)
```

```python
import cv2
import numpy as np
import os

# Ensure directories exist
if not os.path.exists("data"):
    os.makedirs("data")
if not os.path.exists("output"):
    os.makedirs("output")
```

```python
input_image_path = os.path.join("data", "/content/data/bycylce.bmb.webp")
output_gray_image_path = os.path.join("output", "bicyclegray.jpg")
output_binary_image_path = os.path.join("output", "bicycleoutA.jpg")
```

```python
# Check if the file exists
if not os.path.exists(input_image_path):
    print(f"File not found: {input_image_path}")
else:
    print(f"File found: {input_image_path}")
```

```python
image = cv2.imread(input_image_path, cv2.IMREAD_UNCHANGED)
```

```python
if image is None:
    print("Error: Image not loaded. Please check the file path and ensure the file is not corrupted.")
else:
    print("Image loaded successfully.")
```

```python
height = image.shape[0]
width = image.shape[1]
channels = image.shape[2] if len(image.shape) == 3 else 1
image_dtype = image.dtype
num_pixels = image.size
```

Extract
Information

```python
print("Image Height:", height)
print("Image Width:", width)
print("Number of Channels:", channels)
print("Image Datatype:", image_dtype)
print("Number of Pixels:", num_pixels)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite(output_gray_image_path, gray_image)
```

```python
max_value = np.max(gray_image)
mean_value = np.mean(gray_image)
```

```python
print(f"Maximum Pixel Value: {max_value}")
print(f"Mean Pixel Value: {mean_value}")

 # Change pixel values and save
binary_image = np.where(gray_image < mean_value, 0, 1).astype(np.uint8)
cv2.imwrite(output_binary_image_path, binary_image * 255)  # Save binary image
print("The image generated at step (i) is a binary image.")
```
