```python
import pandas as pd

# Replace 'column1', 'column2', etc. with your actual column names


df = pd.read_csv("/content/sample_data/boston_house_prices.csv", header=1)
```

```python
# prompt: make a plot using matplotlib

plt.scatter(df['AGE'], df['TAX'])
plt.xlabel('TAX')
plt.ylabel('AGE')
plt.show()
```

```python
plt.bar(df['AGE'], df['TAX'])
plt.xlabel('TAX')
plt.ylabel('AGE')
plt.style.use('seaborn-whitegrid')
plt.show()
```

```python
# prompt: make a bar plot using only 5 bins

df['AGE'].plot(kind='hist', bins=5)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
df['DIS'].hist(bins=5)

plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: make a box plot do labeliing aslo aslo increase size

df.boxplot()
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=90)
plt.show()
```

```python
# prompt: make 5 type of plot in matplotlib

df.plot(kind='scatter', x='AGE', y='TAX')
df.plot(kind='bar', x='AGE', y='TAX')
df['AGE'].plot(kind='hist', bins=5)
df['DIS'].hist(bins=5)
df.boxplot(rot=45)
```
