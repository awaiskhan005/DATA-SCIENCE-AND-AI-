```python
import plotly.express as px
import pandas as pd

# Sample project data
data = {
    'Project': ['Enroll in Course', 'Seek Internships', 'Attend Conferences', 'Networking Events', 'Update Resume', 'Research Industry'],
    'Start Date': ['2023-02-01', '2024-03-01', '2024-06-01', '2024-08-01', '2024-09-01', '2024-12-01'],
    'Days to complete': [80, 50, 3, 10, 2, 3]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Convert 'Start Date' to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Calculate 'End Date'
df['End Date'] = df['Start Date'] + pd.to_timedelta(df['Days to complete'], unit='D')

# Add a new column for the number of days
df['Days'] = df['Days to complete'].astype(str) + ' days'

# Plot Gantt chart using Plotly
fig = px.timeline(df, x_start='Start Date', x_end='End Date', y='Project', text='Days')
fig.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample project data
data = {
    'Project': ['Enroll in Course', 'Seek Internships', 'Attend Conferences', 'Networking Events', 'Update Resume', 'Research Industry'],
    'Start Date': ['2023-02-01', '2024-03-01', '2024-06-01', '2024-08-01', '2024-09-01', '2024-12-01'],
    'Days to complete': [80, 50, 3, 10, 2, 3]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Convert 'Start Date' to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Calculate 'End Date'
df['End Date'] = df['Start Date'] + pd.to_timedelta(df['Days to complete'], unit='D')

# Prepare data for plotting
df['Duration'] = df['End Date'] - df['Start Date']
df['Start Date'] = df['Start Date'].dt.strftime('%Y-%m-%d')
df = df.melt(id_vars=['Project', 'Start Date', 'End Date'], value_vars=['Duration'], var_name='Type', value_name='Value')

# Plot Gantt chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Start Date', y='Project', hue='Type', dodge=False)
plt.xlabel('Timeline')
plt.ylabel('Tasks')
plt.title('Project Gantt Chart')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample project data
data = {
    'Project': ['Enroll in Course', 'Seek Internships', 'Attend Conferences', 'Networking Events', 'Update Resume', 'Research Industry'],
    'Start Date': ['2023-02-01', '2024-03-01', '2024-06-01', '2024-08-01', '2024-09-01', '2024-12-01'],
    'Days to complete': [80, 50, 3, 10, 2, 3]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Convert 'Start Date' to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Calculate 'End Date'
df['End Date'] = df['Start Date'] + pd.to_timedelta(df['Days to complete'], unit='D')

# Set up the plot
plt.figure(figsize=(10, 6))

# Plot the Gantt chart
for i, row in df.iterrows():
    plt.barh(y=i, width=row['Days to complete'], left=row['Start Date'], color='skyblue', alpha=0.8)
    plt.text(row['Start Date'] + pd.Timedelta(days=row['Days to complete'] / 2), i, f"{row['Project']} ({row['Days to complete']} days)", ha='center', va='center')

# Set labels and title
plt.xlabel('Timeline')
plt.ylabel('Tasks')
plt.title('Project Gantt Chart')

# Invert y-axis to have the tasks listed from top to bottom
plt.gca().invert_yaxis()

# Display the grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample project data
data = {
    'Project': ['Enroll in Course', 'Seek Internships', 'Attend Conferences', 'Networking Events', 'Update Resume', 'Research Industry'],
    'Start Date': ['2023-02-01', '2024-03-01', '2024-06-01', '2024-08-01', '2024-09-01', '2024-12-01'],
    'Days to complete': [80, 50, 3, 10, 2, 3]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Convert 'Start Date' to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])

# Calculate 'End Date'
df['End Date'] = df['Start Date'] + pd.to_timedelta(df['Days to complete'], unit='D')

# Set up the plot
plt.figure(figsize=(10, 6))

# Plot the Gantt chart
for i, row in df.iterrows():
    plt.barh(y=i, width=row['Days to complete'], left=row['Start Date'], color='skyblue', alpha=0.8)
    plt.text(row['Start Date'] + pd.Timedelta(days=row['Days to complete'] / 2), i, f"{row['Project']} ({row['Days to complete']} days)", ha='center', va='center', fontsize=8)

# Set labels and title
plt.xlabel('Timeline')
plt.ylabel('Tasks')
plt.title('Project Gantt Chart')

# Invert y-axis to have the tasks listed from top to bottom
plt.gca().invert_yaxis()

# Display the grid
plt.grid(True)

# Improve readability of date labels on the x-axis
date_ticks = pd.date_range(start=df['Start Date'].min(), end=df['End Date'].max(), freq='1M')
plt.xticks(date_ticks, [date.strftime('%b\n%Y') for date in date_ticks])

# Show the plot
plt.show()
```
