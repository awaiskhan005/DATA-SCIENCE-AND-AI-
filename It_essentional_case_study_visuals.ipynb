```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('/content/sample_data/IT ESSENTIALS.xlsx')
df
```

```python
sns.countplot(data=df, x="Item", hue="Region")
```

```python
# prompt: make plot containing Item , Total, Region using plotly

import plotly.express as px

fig = px.bar(df, x="Region", y="Item", color="Total",
             title="Plot containing Item, Total, Region")
fig.show()
```

```python
# prompt: make a pie chart for item Total using plotly and label the pie aslo

fig = px.pie(df, values='Total', names='Item',
             title='Total per Item',
             labels={'Item':'Item Name'},
             hole=.3)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
# prompt: Use the appropriate graph to compare the number of units of items sold in each region.

sns.barplot(data=df, x="Region", y="Units", hue="Item")
```

```python
# prompt: make a graph that contain good relation ship using plotly

fig = px.scatter(df, x="Total", y="Units", color="Item", trendline="ols")
fig.show()
```
