```python
import sqlite3

# Connect to SQLite
conn = sqlite3.connect("amazon_data.db")
cursor = conn.cursor()

# Create Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS amazon_products (
    product_id TEXT PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    price REAL,
    rating REAL,
    reviews INTEGER,
    stock INTEGER
)
""")

# Insert Sample Data
sample_data = [
    ("P001", "Echo Dot", "Smart Devices", 49.99, 4.5, 12000, 300),
    ("P002", "Fire Stick", "Streaming Devices", 39.99, 4.6, 15000, 500),
    ("P003", "Kindle", "E-Readers", 89.99, 4.7, 8000, 200),
    ("P004", "Alexa Remote", "Accessories", 29.99, 4.2, 5000, 150),
    ("P005", "Amazon Basics Charger", "Electronics", 19.99, 4.1, 2000, 100),
]

cursor.executemany("""
INSERT OR IGNORE INTO amazon_products (product_id, product_name, category, price, rating, reviews, stock)
VALUES (?, ?, ?, ?, ?, ?, ?)
""", sample_data)

conn.commit()
conn.close()
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to SQLite database
engine = create_engine("sqlite:///amazon_data.db")
query = "SELECT * FROM amazon_products"
data = pd.read_sql(query, engine)

# Display data
print(data.head())
```

```python
# Average price and rating
avg_price = data["price"].mean()
avg_rating = data["rating"].mean()

# Top categories by product count
top_categories = data["category"].value_counts()

print(f"Average Price: ${avg_price:.2f}")
print(f"Average Rating: {avg_rating:.2f}")
print("Top Categories:\n", top_categories)
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig = px.bar(
    data,
    x="category",
    y="price",
    color="rating",
    title="Average Price and Rating by Category",
    labels={"price": "Price ($)", "category": "Category"},
    text="rating"
)
fig.show()
```

```python
fig = px.scatter(
    data,
    x="price",
    y="reviews",
    size="stock",
    color="rating",
    hover_name="product_name",
    title="Price vs Reviews with Stock Size",
    labels={"price": "Price ($)", "reviews": "Reviews"},
)
fig.show()
```

```python
import time
from IPython.display import clear_output

for i in range(10):  # Simulate live data updates
    clear_output(wait=True)
    data["price"] += (i * 0.5)  # Simulate price change

    fig = px.line(
        data,
        x="product_name",
        y="price",
        title="Live Price Updates",
        labels={"price": "Price ($)", "product_name": "Product Name"},
    )
    fig.show()
    time.sleep(1)
```

```python
from plotly.subplots import make_subplots

# Create subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=("Price vs Rating", "Stock by Category"))

# Plot 1: Price vs Rating
fig.add_trace(
    go.Scatter(
        x=data["price"],
        y=data["rating"],
        mode="markers",
        marker=dict(size=12, color=data["reviews"], showscale=True),
        name="Price vs Rating"
    ),
    row=1, col=1
)

# Plot 2: Stock by Category
category_stock = data.groupby("category")["stock"].sum().reset_index()
fig.add_trace(
    go.Bar(
        x=category_stock["category"],
        y=category_stock["stock"],
        name="Stock by Category"
    ),
    row=1, col=2
)

# Update Layout
fig.update_layout(
    title="Amazon Product Analysis",
    showlegend=True,
    height=600,
    width=1000
)

fig.show()
```

```python
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)
num_products = 100
data = pd.DataFrame({
    "product_id": [f"P{i:03d}" for i in range(1, num_products + 1)],
    "price": np.random.uniform(10, 500, num_products),
    "rating": np.random.uniform(1, 5, num_products),
    "reviews": np.random.randint(100, 5000, num_products),
    "stock": np.random.randint(50, 1000, num_products),
    "category": np.random.choice(["Electronics", "Books", "Fashion", "Home", "Toys"], num_products)
})
```

```python
import plotly.express as px

fig = px.scatter_3d(
    data,
    x="price", y="rating", z="reviews",
    color="category",
    size="stock",
    hover_name="product_id",
    title="3D Interactive Visualization of Amazon Products",
    labels={"price": "Price ($)", "rating": "Rating", "reviews": "Number of Reviews"},
)

# Customize the 3D view
fig.update_layout(scene=dict(
    xaxis_title="Price ($)",
    yaxis_title="Rating",
    zaxis_title="Reviews",
))

fig.show()
```

```python
import time
from IPython.display import clear_output

# Simulate live updates
for i in range(10):  # 10 updates
    clear_output(wait=True)
    # Simulate price and stock changes
    data["price"] += np.random.uniform(-10, 10, len(data))
    data["stock"] -= np.random.randint(0, 10, len(data))

    # Create 3D scatter plot with updates
    fig = px.scatter_3d(
        data,
        x="price", y="rating", z="reviews",
        color="category",
        size="stock",
        hover_name="product_id",
        title=f"Live 3D Visualization of Amazon Products - Update {i + 1}",
        labels={"price": "Price ($)", "rating": "Rating", "reviews": "Number of Reviews"},
    )
    fig.update_layout(scene=dict(
        xaxis_title="Price ($)",
        yaxis_title="Rating",
        zaxis_title="Reviews",
    ))
    fig.show()
    time.sleep(2)  # Pause for 2 seconds before updating
```

```python
import plotly.graph_objects as go

# Generate dynamic 3D surface data
x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)
x, y = np.meshgrid(x, y)

for i in range(10):  # Simulate updates
    z = np.sin(x + i * 0.1) * np.cos(y + i * 0.1) * 10  # Dynamic surface

    fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])
    fig.update_layout(
        title=f"Dynamic 3D Surface - Frame {i + 1}",
        scene=dict(
            xaxis_title="X Axis",
            yaxis_title="Y Axis",
            zaxis_title="Z Axis",
        ),
    )
    fig.show()
    time.sleep(1)  # Pause for 1 second
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)

# Layout
app.layout = html.Div([
    dcc.Graph(id="live-graph"),
    dcc.Interval(id="interval", interval=2000, n_intervals=0),  # Update every 2 seconds
])

@app.callback(
    Output("live-graph", "figure"),
    Input("interval", "n_intervals")
)
def update_graph(n):
    data["price"] += np.random.uniform(-10, 10, len(data))
    data["stock"] -= np.random.randint(0, 10, len(data))

    fig = px.scatter_3d(
        data,
        x="price", y="rating", z="reviews",
        color="category",
        size="stock",
        hover_name="product_id",
        title=f"Live Amazon Product Visualization (Update {n})"
    )
    return fig

if __name__ == "__main__":
    app.run_server(debug=True)
```

```python
import numpy as np
import pandas as pd

# Generate initial data
np.random.seed(42)
num_products = 50
time_steps = 20

# Simulate product data over multiple time steps
data = []
for t in range(time_steps):
    temp_data = pd.DataFrame({
        "time": t,
        "product_id": [f"P{i:03d}" for i in range(1, num_products + 1)],
        "price": np.random.uniform(10, 500, num_products) + t * np.random.uniform(-5, 5),
        "rating": np.random.uniform(1, 5, num_products),
        "reviews": np.random.randint(100, 5000, num_products) + t * np.random.randint(-20, 20, num_products),
        "stock": np.random.randint(50, 1000, num_products),
        "category": np.random.choice(["Electronics", "Books", "Fashion", "Home", "Toys"], num_products)
    })
    data.append(temp_data)

# Combine into a single DataFrame
animated_data = pd.concat(data)
```

```python
import plotly.express as px

# Create animated scatter plot
fig = px.scatter(
    animated_data,
    x="price", y="rating",
    animation_frame="time",
    animation_group="product_id",
    size="stock",
    color="category",
    hover_name="product_id",
    title="Animated Visualization of Amazon Product Trends Over Time",
    labels={"price": "Price ($)", "rating": "Rating"},
    size_max=40,
)

# Customize layout
fig.update_layout(
    xaxis=dict(title="Price ($)"),
    yaxis=dict(title="Rating"),
    legend_title="Product Category",
)

# Show the animated plot
fig.show()
```
