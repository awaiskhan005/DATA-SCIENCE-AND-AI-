```python
from transformers import BertModel, BertTokenizer, GPT2Model, GPT2Tokenizer
```

```python
bert_model = BertModel.from_pretrained('bert-base-uncased')
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
bloom_model = GPT2Model.from_pretrained('gpt2')
bloom_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
input_text = "Combining BERT and BLOOM models in Google Colab."
```

```python
inputs = bert_tokenizer(input_text, return_tensors="pt")
```

```python
with torch.no_grad():
    bert_output = bert_model(**inputs).last_hidden_state
```

```python
inputs = bloom_tokenizer(input_text, return_tensors="pt")
```

```python
with torch.no_grad():
    bloom_output = bloom_model(**inputs).last_hidden_state
```

```python
combined_output = torch.cat((bert_output, bloom_output), dim=1)

print("Combined Output Shape:", combined_output.shape)
```
