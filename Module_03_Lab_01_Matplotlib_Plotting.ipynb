## 3. Adjusting the spacing around subplots

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

## 5. Ticks, Labels, and Legends

```python
fig = plt.figure() # set for empty figure
ax = fig.add_subplot(1, 1, 1) # create subplot
ax.plot(np.random.randn(1000).cumsum()) # generate randam walk dataset
```

```python
fig = plt.figure() # set for empty figure
ax = fig.add_subplot(1, 1, 1) # create subplot
ax.plot(np.random.randn(1000).cumsum()) # generate randam walk dataset
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],rotation=45, fontsize='small')
ax.set_title("Dost's first matplotlib plot")
ax.set_xlabel('Stages')
```
