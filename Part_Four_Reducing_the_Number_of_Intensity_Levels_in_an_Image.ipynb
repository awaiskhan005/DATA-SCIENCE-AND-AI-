Reducing the Number of Intensity Levels in an Image

Introduction:
Image processing is a critical aspect of various fields, including computer vision, medical imaging, and satellite imagery analysis. One common task in image processing is reducing the number of intensity levels in an image, which can help in compressing images and reducing computational complexity. In this report, we explore the process of reducing intensity levels in a grayscale image from 256 to 2 levels, in integer powers of 2.

Objective:
The objective of this report is to develop a computer program capable of reducing the intensity levels in a grayscale image and saving the modified images with the required intensity levels.

Methodology:

Loading the Image: The process begins by loading the input image "lena.tif" from the data directory. The image is loaded using the Python Imaging Library (PIL).

Converting to Grayscale: If the image is not already in grayscale, it is converted to grayscale using the 'convert' function in PIL.

Reducing Intensity Levels: The intensity levels in the grayscale image are reduced from 256 to 2 levels, in integer powers of 2 (i.e., 256, 128, 64, 32, 16, 8, 4, 2). This is achieved by dividing the maximum intensity value by the desired number of levels and rounding the pixel values accordingly.

Saving Modified Images: Each modified image is saved with the appropriate filename and extension in the output directory. The filenames follow the format "lena[level].jpg" for levels greater than or equal to 16 and "lena[level].png" for levels less than 16.

Results:
The developed computer program successfully reduced the intensity levels in the input image "lena.tif" to the desired levels. The modified images were saved in the output directory with the correct filenames and extensions. The generated images allow for visual comparison of the original and modified intensity levels.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from PIL import Image
import numpy as np
import os

# Define the input and output directories
input_dir = "/path/to/input/directory"
output_dir = "/path/to/output/directory"
```

```python
# Load the image
image_path = os.path.join(input_dir, "/content/sample_data/unnamed.jpg")
lena_image = Image.open(image_path)
```

```python


plt.imshow(lena_image)
plt.show()
```

```python
# Convert the image to grayscale
lena_gray = lena_image.convert('L')
```

```python
plt.imshow(lena_gray, cmap='gray')
plt.show()
```

```python
# Define intensity levels
intensity_levels = [256, 128, 64, 32, 16, 8, 4, 2]
```

```python
# Reduce intensity levels and save images
for level in intensity_levels:
    # Convert image to numpy array
    img_array = np.array(lena_gray)

    # Reduce intensity levels
    max_val = np.max(img_array)
    factor = max_val / (level - 1)
    reduced_array = (img_array / factor).astype(int) * factor
```

```python
reduced_image = Image.fromarray(reduced_array.astype(np.uint8))
output_filename = f"lena{level}.jpg" if level <= 16 else f"lena{level}.png"
output_path = os.path.join(output_dir, output_filename)
# Save the image
reduced_image.save(output_path)

print("Images with reduced intensity levels have been saved successfully.")
```

```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt

# Define the input and output directories
input_dir = "/path/to/input/directory"
output_dir = "/path/to/output/directory"

# Load the image
image_path = os.path.join(input_dir, "/content/sample_data/unnamed.jpg")
lena_image = Image.open(image_path)

# Convert the image to grayscale
lena_gray = lena_image.convert('L')

# Define intensity levels
intensity_levels = [256, 128, 64, 32, 16, 8, 4, 2]

# Reduce intensity levels and save images
for level in intensity_levels:
    # Convert image to numpy array
    img_array = np.array(lena_gray)

    # Reduce intensity levels
    max_val = np.max(img_array)
    factor = max_val / (level - 1)
    reduced_array = (img_array / factor).astype(int) * factor

    # Convert numpy array back to image
    reduced_image = Image.fromarray(reduced_array.astype(np.uint8))

    # Define output filename with appropriate extension
    output_extension = "jpg" if level >= 16 else "png"
    output_filename = f"lena{level}.{output_extension}"
    output_path = os.path.join(output_dir, output_filename)

    # Save the image
    reduced_image.save(output_path)

    # Display the image
    plt.imshow(reduced_image, cmap='gray')
    plt.title(f"Lena {level} intensity levels")
    plt.axis('off')
    plt.show()

print("Images with reduced intensity levels have been displayed successfully.")
```
